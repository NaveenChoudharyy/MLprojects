###### For creating movies recommendation system we need to find the similarity between the movies, so we required only those variables which can provide any information if that movie is similar with another ovie or not.

```python
movies = movies[['genres', 'id', 'keywords', 'original_title', 'overview', 'spoken_languages', 'tagline', 'title']]

movies.head()
```

```python
movies.isna().sum()

# We need null value tratment as there are some missings
```

```python
# We need to drop the records where overview is itself missing as it contains lots of information 
# about the sismilarity of the movie which cannot be treated or replaced.

# Whereever "tagline" is missing we can replace null with blank ("") as it doesnot contain that much information
# and this will avoid the information loss also.


movies = movies.dropna(subset="overview")

movies = movies.fillna(value="")

movies.isna().sum()
```

```python
import json
```

```python
# Extracting required information from variable "genres"

movies.genres.apply( lambda x: json.loads(x) )[1]
```

```python
# From variable gerne we need only "name"

movies.genres = movies.genres.apply( lambda x: [genres['name'] for genres in eval(x)])


movies.head()
```

```python
# Likewise we have to do this for all columns


# Extracting required information from variable "genres"

json.loads(movies.loc[1, 'keywords'])
```

```python
movies.keywords = movies.keywords.apply( lambda x: [keywords['name'] for keywords in eval(x)] )


movies.head()
```

```python
# Extracting required information from variable "spoken_languages"


movies.loc[1, 'spoken_languages']
```

```python
movies.spoken_languages = movies.spoken_languages.apply( lambda x: 
                                                        [spoken_languages['name'] for spoken_languages in eval(x)] )


movies.head()
```

### Credits

```python
credits.head()
```

```python
credits.info()

# No nulls available
```

```python
# Extracting required information from variable "cast"


json.loads(credits.loc[1, 'cast'])

# Here, we need to extract both "character" and "name"
```

```python
credits.cast = credits.cast.apply( lambda x: [cast['character'] for cast in 
                                              eval(x)]) + credits.cast.apply( lambda x: 
                                                                             [cast['name'] for cast in eval(x)])


credits.head()
```

```python
# Extracting required information from variable "crew"


json.loads(credits.loc[1, 'crew'])


# Here, we need only name to be extracted
```

```python
credits.crew = credits.crew.apply( lambda x: [crew['name'] for crew in eval(x)] )


credits.head()
```

```python
credits
```

```python
# Mergeing both the datasets


df = pd.merge(left=movies, right=credits, left_on=['id', 'title'], right_on=['movie_id', 'title'], how='inner')
```

```python
df
```

```python
df.info()
```

```python
import re
```

```python
# Tokenization for variable "overview"

df['overview'] = df['overview'].apply(lambda x: re.split(r'[ :`~#!@$%\^&*()\[\]{}\-_+=;\"<,.>/?|]', x))

df['tagline'] = df['tagline'].apply(lambda x: re.split(r'[ :`~#!@$%\^&*()\[\]{}\-_+=;\"<,.>/?|]', x))
```

```python
import string
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from wordcloud import STOPWORDS

stp_words = list(set(list(ENGLISH_STOP_WORDS) + list(STOPWORDS)))
```

```python
# Removing stopwords, single charaters and panctuations

df['overview'] = df.overview.apply( lambda x: [w for w in x if 
                                               len(w) > 1 and 
                                               w not in stp_words and 
                                               w not in string.punctuation] )


df['tagline'] = df.tagline.apply( lambda x: [w for w in x if 
                                               len(w) > 1 and 
                                               w not in stp_words and 
                                               w not in string.punctuation] )
```

```python
df = df[['movie_id', 'title', 'overview', 'tagline', 'genres', 'keywords', 'spoken_languages', 'cast', 'crew']]
```

```python
df['merged_col_1'] = df.overview + df.tagline + df.keywords


df = df.drop(columns=['overview' , 'tagline' , 'keywords'])


df.head()
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from textblob import TextBlob
```

```python
df.merged_col_1 = df.merged_col_1.apply( lambda x: " ".join(x))


df.merged_col_1 = df.merged_col_1.apply( lambda x: WordNetLemmatizer().lemmatize(x) )


df.head()
```

```python
df['merged_col_2'] = df.genres + df.spoken_languages + df.cast + df.crew

df.merged_col_2 = df.merged_col_2.apply( lambda x: " ".join(x))

df.head()
```

```python
df = df.drop(columns=['genres','spoken_languages','cast','crew'])

df.head()
```

```python
df['merged_col'] = df.merged_col_1 + ' ' + df.merged_col_2

df = df.drop(columns=['merged_col_1', 'merged_col_2'])

df.head()
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cv = CountVectorizer(token_pattern=r'\b[a-zA-Z]+\b', max_features=10000)
```

```python
X = cv.fit_transform(df.merged_col)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

X = cosine_similarity(X)
```

```python
X
```

```python
def recommend_movie(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = X[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend_movie('Liar Liar')
```
